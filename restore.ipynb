```python
# Read csv file 
import csv

video_ids = []
with open('small_sample.csv', mode ='r')as file:
	csvFile = csv.reader(file)
	for i, lines in enumerate(csvFile):
		if i > 0:
			video_ids.append(lines[0])

```

```python
with open('restore_ids.txt','w') as file:
    for video_id in video_ids:
    	file.write(f"{video_id}\n")
```
